```python
imlist = os.listdir("img")
```

```python
for n in range(len(pl)):
    p = pl[n]
    sn = (p['name'].replace(" ", "")).lower()
    impath = False
    for im in imlist:
        #print(im)
        #print(sn)
        if sn in im:
            impath = im 
    if impath != False:
        image = "../assets/img/"+impath
    else:
        image = "../assets/img/portrait_placeholder.png"
    p['image'] = image
    if 'categories' in p:
        if ['committee'] in p['categories']:
            p['categories'] = ['committee']
    else:    
        p['categories'] = []
    if 'social' in p:
        p['links'] =[]
        if 'github' in p['social'] :
            #p['links'] =  '<a href=https://github.com/'+ str(p['social']['github'])+ '><i class=\"fab fa-github\"></i></a>'+'<a href='+str(p['social']['website'])+'><i class=\"fas fa-globe\"></i></a>'
            p['links'] += ['<a href=https://github.com/'+ str(p['social']['github'])+ ' class="btn btn-primary">Github</a>']
        if 'website' in p['social']:
            #p['links'] =  '<a href='+str(p['social']['website'])+'><i class=\"fas fa-globe\"></i></a>'
            p['links'] += ['<a href='+ str(p['social']['website'])+ ' class="btn btn-secondary">Website</a>']
    if 'neural' in  p['techniques'] or 'CNN' in  p['techniques'] or 'cnn' in  p['techniques'] or 'Cnn' in  p['techniques'] or 'deep learning' in  p['techniques']:
        p['categories']+= ['Deep Learning']
    if p['techniques'] in  ('CNN','cnn','convolu') #or 'cnn' in  p['techniques'] or 'Cnn' in  p['techniques'] or 'convolutational' in  p['techniques']:
        p['categories']+= ['CNN']
    if 'forest' in  p['techniques'] or 'boost' in  p['techniques'] or 'Boost' in  p['techniques'] or 'Forest' in  p['techniques']:
        p['categories']+= ['Decision Trees']

    if 'PINN' in  p['techniques'] or 'pinn' in  p['techniques'] or 'physics informed' in  p['techniques']:
        p['categories']+= ['PINNS']
    pl[n] = p
    
```

```python
with open('people_processed.yml', 'w') as outfile:
    yaml.dump(pl, outfile, default_flow_style=False)
```
